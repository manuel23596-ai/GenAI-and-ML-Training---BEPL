```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# --- 1. Load the Dataset ---
print("--- Loading Iris Dataset ---")
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target_names[iris.target] # Get target names for better labeling

# Create a DataFrame that includes the species label (for easier plotting)
df = X.copy()
df['species'] = y

print("Dataset loaded successfully.")
print("\n--- First 5 Rows of the Dataset ---")
print(df.head())


# ===============================================================
# --- 2. Exploratory Data Analysis (EDA) Visualizations ---
# ===============================================================

# Set a professional style for the plots
sns.set_style("whitegrid")

# --- A. Pair Plot (Relationship between all features across species) ---
print("\n--- Generating Pair Plot (Relationships between Features) ---")
# pairplot creates its own figure, so we handle it first.
pair_fig = sns.pairplot(df, hue='species', markers=["o", "s", "^"])
plt.suptitle('Pairwise Relationships of Iris Features by Species', fontsize=16, y=1.02)
plt.show() # Show the pair plot immediately


# --- B. Distribution & Comparison Plots (Using a dedicated figure for clarity) ---
print("\n--- Generating Feature Distributions and Comparisons ---")

# We need 3 rows: Row 1 for Histograms, Row 2 for Box Plots, Row 3 (optional/empty)
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(14, 15)) # 2 Rows total
plt.suptitle('Feature Distribution and Comparison Analysis', fontsize=18, y=1.02)


# --- PLOT ROW 1: Histograms (Distribution Shape) ---
print("Plotting Feature Distributions...")
for i, feature in enumerate(iris.feature_names):
    ax = axes[0] # Use the first row for all histograms
    sns.histplot(df[feature], kde=True, ax=axes[0])
    # Since we are plotting 4 features on one axis, we must use subplots within this row if possible,
    # but to keep it simple and fix the error, let's plot them sequentially in a single figure structure.

# *** REVISING THE SUBPLOT STRUCTURE FOR SIMPLICITY AND CORRECTNESS ***
# Instead of trying to cram everything into one complex subplot grid, we will use separate figures for clarity.


# --- NEW APPROACH: Separate Figures for Different Plot Types ---

# 1. Histograms (One figure showing all distributions)
fig_hist, axes_hist = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))
axes_hist = axes_hist.flatten() # Flatten the 2x2 array of axes for easy iteration

for i, feature in enumerate(iris.feature_names):
    ax = axes_hist[i]
    sns.histplot(df[feature], kde=True, ax=ax)
    ax.set_title(f'Distribution of {feature}', fontsize=12)
    ax.set_xlabel("") # Clear x-label for cleaner look

plt.suptitle('Feature Distribution Analysis (Histograms)', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()


# 2. Box Plots (One figure comparing species across all features)
fig_box, axes_box = plt.subplots(nrows=3, ncols=2, figsize=(14, 15)) # 3 Rows x 2 Cols
axes_box = axes_box.flatten()

for i, feature in enumerate(iris.feature_names):
    ax = axes_box[i]
    sns.boxplot(x='species', y=feature, data=df, ax=ax)
    ax.set_title(f'Box Plot of {feature}', fontsize=12)

plt.suptitle('Feature Comparison Across Species (Box Plots)', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()


print("\n=============================================================")
print("EDA Complete! Three separate figures were generated for:")
print("1. Pair Plot: Feature relationships.")
print("2. Histograms: Individual feature distributions.")
print("3. Box Plots: Comparison of features across the three species.")

```
